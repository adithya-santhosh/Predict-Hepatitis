```python
hepatitis_df_modified = hepatitis_df["Steroid"].replace("?",0)
```

```python
replace_list = ["Steroid", "Fatigue", "Malaise", "Anorexia", "Liver Big", "Liver Firm", "Spleen Palpable", "Spiders", "Ascities", "Varices"]
mean_list = ["Bilirubin","ALK Poshphate", "SGOT", "Albumin", "Protime", "Histology"]
hepatitis_df_modified.head(10)
```

```python
df_removed = hepatitis_df.drop(hepatitis_df[hepatitis_df["Fatigue"]=="?"].index)
for i in replace_list:
    df_removed = df_removed.drop(df_removed[df_removed[i]=="?"].index)
for j in mean_list:
    temp_data = df_removed.drop(df_removed[df_removed[j]=="?"].index)
    temp_data[j] = pd.to_numeric(temp_data[j])
    df_removed = df_removed.replace("?", temp_data[j].mean())
df_removed
```

```python
df_removed.to_csv("test_csvfile.csv")
```
