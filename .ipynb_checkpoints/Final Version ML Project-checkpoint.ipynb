## Handiling missing data
We're going to remove the null values for certain columns and for other we are going to use mean

```python
modified_data = data
replace_list = ["Steroid", "Fatigue", "Malaise", "Anorexia", "Liver Big", "Liver Firm", "Spleen Palpable", "Spiders", "Ascities", "Varices"]
mean_list = ["Bilirubin","ALK Poshphate", "SGOT", "Albumin", "Protime", "Histology"]

for i in mean_list:
    modified_data[i] = modified_data[i].fillna(modified_data[i].mean())
modified_data.dropna(inplace=True)
modified_data
```

```python
# to confirm that no Null values are there.
modified_data.isnull().sum()
```

```python
# Feature scaling
minmax = MinMaxScaler(feature_range=(0,1))
modified_data["ALK Poshphate"]= minmax.fit_transform(modified_data["ALK Poshphate"].values.reshape(-1,1))
modified_data["Protime"]= minmax.fit_transform(modified_data["Protime"].values.reshape(-1,1))
modified_data.head()
```

```python
x = modified_data.drop(['Steroid'], axis=1)
y = modified_data['Steroid']
sm = SMOTE(random_state=42, k_neighbors=5)
X_res, y_res = sm.fit_resample(x,y)
X_res
```

## Spliting the data and Fitting the Model

```python
# Here we split the dataset into two data and target
X = modified_data.drop(columns=["Unnamed: 0", "Class"])
y = modified_data["Class"]
```

### `KNN Classifier(K Neighbours)`

```python
# Splitting Of Dataset into training and testing
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size = 0.3,random_state=42)

# Using KNeighborClassifier with certain parameter
knn = KNeighborsClassifier(n_neighbors=3)

# Fitting(Training) the model with training dataset (x_train, y_train)
knn.fit(x_train,y_train)

# Predicting the output of x_test
y_pred1 = knn.predict(x_test.values)
y_pred1
```
